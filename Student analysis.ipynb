# Visualizations

#Histogram Of Grades

```python
plt.figure(figsize=(8,5))
plt.hist([df['G1'], df['G2'], df['G3']], label=['G1','G2','G3'])
plt.title("Histogram of All Grades")
plt.xlabel("Grades")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

#The grade distribution for G1, G2, and G3 shows a normal-like pattern centered around 10â€“13, with gradual improvement 
toward G3 and a small number of very high or very low achievers.

#Scatterplot: study time vs grades.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,5))

plt.scatter(df['studytime'], df['G1'], label='G1')
plt.scatter(df['studytime'], df['G2'], label='G2')
plt.scatter(df['studytime'], df['G3'], label='G3')

plt.xlabel("Study Time")
plt.ylabel("Grades")
plt.title("Study Time vs All Grades (G1, G2, G3)")
plt.legend()

plt.show()
```

#There is a positive relationship between study time and grades, although not perfectly linear.
Students who dedicate more time to studying tend to achieve higher grades across G1, G2, and G3.
